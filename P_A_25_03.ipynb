```python
#Escrever um programa que recebe 04 notas do aluno, calcula a média e mostra se o aluno foi aprovado

#Entrada

print(f"Digite suas notas aqui de 01 a 10")
nota1 = float(input("Digite a nota do 1º bimestre: "))
nota2 = float(input("Digite a nota do 2º bimestre: "))
nota3 = float(input("Digite a nota do 3º bimestre: "))
nota4 = float(input("Digite a nota do 4º bimestre: "))

#Processamento

total = nota1 + nota2 + nota3 + nota4
media = total / 4

#Saída


if media >= 7:
  print(f"A média das suas notas foi {media}, você foi aprovado")

elif media >= 5:
  print(f"A média das suas notas foi {media}, você está de recuperação")

else :
  print(f"A média das suas notas foi {media}, você foi reprovado")
```
