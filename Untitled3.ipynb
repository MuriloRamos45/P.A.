```python
#Escrever um programa que recebe 4 notas referentes aos bimestres, calcula a média e mostra se o aluno foi aprovado, recuperação ou reprovado

nome = str(input("Digite o nome do aluno: "))
print("")
a = float(input("Digite a nota do 1ºBimestre: "))
if a > 10:
  print("Digite uma nota válida (de 0 a 10)")
b = float(input("Digite a nota do 2ºBimestre: "))
if b > 10:
  print("Digite uma nota válida (de 0 a 10)")
c = float(input("Digite a nota do 3ºBimestre: "))
if c > 10:
  print("Digite uma nota válida (de 0 a 10)")
d = float(input("Digite a nota do 4ºBimestre: "))
if d > 10:
  print("Digite uma nota válida (de 0 a 10)")

#Processamento

media = (a + b + c + d) / 4

#Saída
print("")
print(f"A média do aluno {nome} foi {media}")
print("")


if media >= 7:
  print(f"O aluno foi aprovado")

elif media >=5:
  print(f"O aluno ficou de recuperação")

else:
  print(f"O aluno foi reprovado")


```
