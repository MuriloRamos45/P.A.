Escrever um programa que receba o nome do usuário e a idade e mostre uma mensagem com o ano que ele nasceu



```python
#Entrada

nome = str (input("Digite seu nome " ))
idade = int (input("Digite sua idade " ))

#Processamento

nasceu = 2026 - idade

#Saída

print(f"Bem Vindo {nome} ")
print(f"Sua idade é {idade} anos ")
print(f"Você nasceu em {nasceu} ")
```
