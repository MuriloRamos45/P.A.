```python
# Operadores aritiméticos
#Entrada

a = int(input("digite um número: "))
b = int(input("digite um número: "))

#Processamento

soma = a + b
subrtacao = a - b
multiplicacao = a * b
divisao = a / b
diviint = a//b
resto = a % b
potencia = a ** b


#saída
print("")
print("----------------------------Operadores Aritiméticos----------------------------")
print("")
print("-------------------------------------Soma--------------------------------------")
print(f"A soma de {a} mais {b} é igual a:{soma}")
print("")
print("------------------------------------Subtração--------------------------------------")
print(f"A subtração de {a} menos {b} é igual a:{subrtacao}")
print("")
print("-----------------------------------Multplicação--------------------------------------")
print(f"A multiplicação de {a} vezes {b} é igual a:{multiplicacao}")
print("")
print("-------------------------------------Divisão--------------------------------------")
print(f"A divisão de {a} dividido por {b} é igual a:{divisao}")
print("")
print("---------------------------------Divisão-Inteira-------------------------------------")
print(f"A divisão inteira de {a} dividido por {b} é igual a:{diviint}")
print("")
print("-----------------------------------Resto-Divisão--------------------------------------")
print(f"O resto da divisão de {a} dividido por {b} é igual a:{resto}")
print("")
print("-------------------------------------Potenciação--------------------------------------")
print(f"A potenciação de {a} elevado a {b} é igual a:{potencia}")
print("")
```
